```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_path = 'dadosImagem/Treinamento/positivos/crop_000010.png'
img_test = cv2.imread(img_path) 
plt.imshow(cv2.cvtColor(img_test,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_test.shape
```

```python
print(np.min(img_test))
print(np.max(img_test))
```

```python
img_test_gray = cv2.cvtColor(img_test, cv2.COLOR_RGB2GRAY)
plt.imshow(img_test_gray,cmap='gray')
plt.show()
```

```python
print(img_test_gray.shape)
print(np.min(img_test_gray))
print(np.max(img_test_gray))
```

```python
img_resized = cv2.resize(img_test_gray,(360,360), interpolation=cv2.INTER_CUBIC)
plt.subplot(121)
plt.title('Original')
plt.imshow(img_test_gray,cmap='gray',interpolation='bicubic')
plt.subplot(122)
plt.title('Resized')
plt.imshow(img_resized,cmap='gray',interpolation='bicubic')
plt.show()
```

```python
plt.imshow(img_resized,cmap='gray',interpolation='bicubic')
plt.show()
```

```python
histogram_analise = cv2.calcHist([img_resized],[0],None,[256],[0,256]).astype(np.int)
plt.plot(histogram_analise)
plt.show()
```

```python
img_test_equalize = cv2.equalizeHist(img_resized)
histogram_equalize = cv2.calcHist([img_test_equalize],[0],None,[256],[0,256]).astype(np.int)
plt.plot(histogram_equalize)
plt.show()
```
